Import library

```python
import numpy as np
import os
import pickle
from kmeans import Member, Cluster, Kmeans
```

Load data

```python
kmeans = Kmeans(num_clusters=8)
kmeans.load_data("../../Session 1/source/20news-bydate/data_tf_idf.txt")
```

Run Kmeans

```python
kmeans.run(seed_value=202, criterion="max_iters", threshold=6)
pickle.dump(kmeans, open("../source/kmeans.sav", 'wb'))
```

```python
loaded_model = pickle.load(open("../source/kmeans.sav", 'rb'))
print("Purity = {}".format(loaded_model.compute_purity()))
```
